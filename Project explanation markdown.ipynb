## !. What is the project?
We aimed to make a french teaching tool that helps beguinner level students get a grasp of french language and culture. for this purporse we required an agent that would do the following:
1. Make a curriculum to follow
2. Teach the material
3. Quiz the student
4. Evaluate the results of the quiz
5. Motivate the learner
6. Save the progress and increase difficulty progessively
We made 2 models one a single agent system and the other a multiple agent in gemini ai studio and model gemini flash 2.5. The multiple agent system was divided by the following roles:
# planner
Creates topic, objectives, vocabulary, culture notes.
It acts in the background.

# explainer
Adds simple explanations + pronunciation tips.

## 2. Advantages of the multi agent system over the single agent system
The major benefit of the multi agent system is modularity. Since we are able to divide the roles, the system was easier to design, debug and extend. The multi agent system also runs faster since it allows for parallel processing allowing more than one agent to act on different parts of a problem. Updating the system is also easier since we can replace or improve one agent without interfering with the others. Lastly, the separation of concerns allowed for better control of lesson generation, adaptive feedback and safety monitoring.

## 3. What are the challenges faced in this project?
We had an issue handling token effeciency. Originally we used the free tier of gemini 3 preview, the free tokens provided were very limited and so we shifted to gemini 2.5 flash that offered more. This necessitated careful prompt optimization, reduction of nested structures and limiting stored information to essential variables. Another difficulty involved maintaining consistent state transitions across phases, especially when adding navigation commands and human-in-the-loop mechanisms.
